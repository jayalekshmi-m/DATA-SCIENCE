```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# 1️⃣ Load the dataset
df = pd.read_csv("naive_bayes.csv")

# 2️⃣ Create a separate LabelEncoder for each column
le_swim = LabelEncoder()
le_fly = LabelEncoder()
le_crawl = LabelEncoder()
le_class = LabelEncoder()

df['swim'] = le_swim.fit_transform(df['swim'])
df['fly'] = le_fly.fit_transform(df['fly'])
df['crawl'] = le_crawl.fit_transform(df['crawl'])
df['class'] = le_class.fit_transform(df['class'])

# 3️⃣ Separate features and target
X = df[['swim', 'fly', 'crawl']]
y = df['class']

# 4️⃣ Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5️⃣ Train Naive Bayes model
model = GaussianNB()
model.fit(X_train, y_train)

# 6️⃣ Predict test data
y_pred = model.predict(X_test)

# 7️⃣ Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# 8️⃣ Predict a new sample
# Example: swim="fast", fly="no", crawl="no"
new_sample = [[
    le_swim.transform(["fast"])[0],
    le_fly.transform(["no"])[0],
    le_crawl.transform(["no"])[0]
]]

prediction = model.predict(new_sample)
print("Predicted Class:", le_class.inverse_transform(prediction)[0])
```

```python
# another ques based on diabetes dataset

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.naive_bayes import GaussianNB
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df=pd.read_csv("sample2.csv")

x=df[["hours_studied","attendance","prev_grade"]]
y=df["Passed"]

le=LabelEncoder()
x['prev_grade']=le.fit_transform(x["prev_grade"])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

model=GaussianNB()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

accuracy=accuracy_score(y_test,y_pred)
classification=classification_report(y_test,y_pred)
print(accuracy)
print(classification)

new=[[6,90,le.transform(['b'])[0]],[2,55,le.transform(['d'])[0]]]
pred=model.predict(new)
print("predicted result",pred)
print("predicted result:",le.inverse_transform(pred)[0])

```

```python
# to make result as passed or failed instead of 1 0

# Mapping numeric output to text labels
result_map = {0: "Failed", 1: "Passed"}

for sample, pred in zip(new, pred):
    print(f"Input {sample} → {result_map[pred]}")
```
