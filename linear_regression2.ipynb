```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python

# Load Dataset
df = pd.read_csv("linear_regression.csv")

# Selecting independent and dependent variables
X = df[["YearsExperience"]]
y = df[["Salary"]]

# Splitting into Training and Testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creating and training model
model = LinearRegression()
model.fit(X_train, y_train)

# Predicting values
y_pred = model.predict(X_test)

# =======================
# 1) Plot for Training Data
# =======================

plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, color='red')         # Training points
plt.plot(X_train, model.predict(X_train), color='blue')  # Regression line
plt.title("Salary vs Experience (Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid(True)
plt.show()

# =======================
# 2) Plot for Testing Data
# =======================

plt.figure(figsize=(6,4))
plt.scatter(X_test, y_test, color='red')           # Testing points
plt.plot(X_test, y_pred, color='blue')             # Regression line
plt.title("Salary vs Experience (Testing Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid(True)
plt.show()

# =======================
# 3) Actual vs Predicted Comparison
# =======================

plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, color='blue', edgecolors='k', alpha=0.7)  # Comparison
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', linewidth=1.5)  # Perfect prediction line
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary (Model Accuracy Check)")
plt.grid(True)
plt.show()
```

```python
# =======================
# 4) Predict Salary for New Data
# =======================

new_exp = float(input("Enter Years of Experience: "))   # user input
new_salary_pred = model.predict([[new_exp]])            # prediction

print("\nPredicted Salary for", new_exp, "years of experience = ", float(new_salary_pred))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Coefficient (Slope)
print("Coefficient (m):", model.coef_)

# Intercept (b)
print("Intercept (b):", model.intercept_)

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# R^2 Score
r2 = r2_score(y_test, y_pred)
print("R^2 Score:", r2)
```
