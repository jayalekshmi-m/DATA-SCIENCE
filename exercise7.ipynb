```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 2: Feature and target names
fname = iris.feature_names
tname = iris.target_names

# Step 3: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Create Decision Tree model (using entropy)
clf = DecisionTreeClassifier(criterion="entropy")
clf.fit(X_train, y_train)

# Step 5: Predict a new case
newcase = [[0, 0, 1, 0]]
y_new = clf.predict(newcase)
print("Predicted class for new case:", tname[y_new[0]])

# Step 6: Predict using test data
y_pred = clf.predict(X_test)

# Step 7: Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Step 8: Visualize the Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(clf,  filled=True, rounded=True,
          class_names=tname, feature_names=fname)
plt.title("Decision Tree Visualization (Iris Dataset)")
plt.show()
```

```python
#explanation for step 8:
# plot_tree() draws the actual decision tree.
# filled=True  colors nodes by predicted class.
# rounded=True  makes nodes rounder (for style).
# class_names and feature_names label the tree with readable names.
```
