```python
import pandas as pd
import matplotlib.pyplot as plt   
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression  
from sklearn.model_selection import train_test_split
```

```python
# 2️⃣ Load the iris dataset
iris = load_iris()
x = iris.data[:, :2]  # Take first two columns: sepal length & sepal width
y = iris.data[:, 2]   # Take third column: petal length (this is what we want to predict)

# 3️⃣ Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.3, random_state=42)

# 4️⃣ Create and train the linear regression model
model = LinearRegression()
model.fit(x_train, y_train)

# 5️⃣ Predict petal lengths for test data
y_pred = model.predict(x_test)

# 6️⃣ Plot actual vs predicted
plt.scatter(y_test, y_pred, edgecolor="k", alpha=0.7)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--', linewidth=2)
plt.xlabel("Actual Petal Length (cm)")
plt.ylabel("Predicted Petal Length (cm)")
plt.title("Linear Regression: Actual vs Predicted Petal Length")
plt.show()
```

```python
#explanation for step6 actual vs predicted

# plt.scatter(y_test, y_pred) → plot each test point: x = actual value, y = predicted value
# edgecolor="k" → black border around dots
# alpha=0.7 → makes dots slightly transparent
# plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--', linewidth=2) → plots a red dashed line representing y = x (perfect prediction)
# This line helps see how close predictions are to actual values.
# ✅ If dots lie on the red dashed line, predictions are perfect.
#The red dashed line is a reference for perfect prediction.
# Interpretation:
# On the line → perfect prediction
# Above the line → model predicted higher than actual
# Below the line → model predicted lower than actual
```
