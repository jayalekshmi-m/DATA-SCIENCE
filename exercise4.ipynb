```python
listdata=[
    [1,"alice",15],
    [2,"bob",15],
    [3,"don",45]
]
df=pd.DataFrame(listdata,columns=['id','name','score'])
print("data frame")
print(df)
# sorting first by score then by ascending order of name 
sorted_df=df.sort_values(by=['score','name'],ascending=[False,True])
print("sorted dataframe")
print(sorted_df)
```

```python
data=[
    ['alice','teacher',5500],
    ['bob','doctor',6000],
    ['emma','doctor',7000],
    ['charlie','techer',2000],
    ['dylan','enginner',3500]
]
df=pd.DataFrame(data,columns=['name','job','salary'])
print(df)
avg_by_job=df.groupby('job')['salary'].mean()
print(avg_by_job)
```

```python
data1=[
    [101,'john',30000],
    [102,'amith',35000],
    [103,'sona',20000]
]
df1=pd.DataFrame(data1,columns=['id','name','salary'])

data2=[
    [101,'developer'],
    [102,'tester'],
    [105,'engineer']
]
df2=pd.DataFrame(data2,columns=['id','position'])

df3=pd.merge(df1,df2,how='inner',on='id')
print(df3)
```
