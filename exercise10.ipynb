```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# 1️⃣ Load the iris dataset
iris = load_iris()
x = iris.data[:, :2]  # use first two features: sepal length & sepal width
y = iris.data[:, 2]   # target: petal length

# 2️⃣ Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# 3️⃣ Create and train linear regression model
model = LinearRegression()
model.fit(x_train, y_train)

# 4️⃣ Predict using the trained model
y_pred = model.predict(x_test)


# 5️⃣ Plot setup
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")


# 6️⃣ Scatter plot for actual data
ax.scatter(x_test[:, 0], x_test[:, 1], y_test, color="blue", label="Actual")


# 7️⃣ Create a mesh (grid) surface to visualize prediction plane
x_surf, y_surf = np.meshgrid(
    np.linspace(x[:, 0].min(), x[:, 0].max(), 20),
    np.linspace(x[:, 1].min(), x[:, 1].max(), 20)
)


# Calculate predicted z values (the plane)
z_surf = model.intercept_ + model.coef_[0] * x_surf + model.coef_[1] * y_surf


# 9️⃣ Label axes and title
ax.set_xlabel("Sepal Length (cm)")
ax.set_ylabel("Sepal Width (cm)")
ax.set_zlabel("Petal Length (cm)")
ax.set_title("3D Linear Regression Plane")

plt.legend()
plt.show()
```
