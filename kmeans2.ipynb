```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("kmeans.csv")
print(df)
```

```python
x=df[["annual_income","spending_score"]]
y=df["age"]

kmeans=KMeans(n_clusters=3,random_state=21)
kmeans.fit(x)
labels=kmeans.labels_

plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
plt.scatter(x["annual_income"],x["spending_score"],c=y,cmap="prism",edgecolors="k")
plt.title("actual")
plt.xlabel("annual")
plt.ylabel("spending")

plt.subplot(1,2,2)
plt.scatter(x["annual_income"],x["spending_score"],c=labels,cmap="jet",edgecolors="k")
plt.title("predicted")
plt.xlabel("annual")
plt.ylabel("spending")

plt.tight_layout()
```

```python
#another ques

from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("sample4.csv")

x=df[['age','annual_income','gender']]
y=df.iloc[:,-1]

le=LabelEncoder()
x['gender']=le.fit_transform(x['gender'])
y=le.fit_transform(y)

model=KMeans(n_clusters=3)
model.fit(x)
labels=model.labels_

print("cluster center:",model.cluster_centers_)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(x['age'],x['annual_income'],c=y,cmap='prism')
plt.xlabel("age")
plt.ylabel("income")
plt.title("actual classes")

plt.subplot(1,2,2)
plt.scatter(x['age'],x['annual_income'],c=labels,cmap='jet')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],marker="*",s=200,color="black")
plt.xlabel("age")
plt.ylabel("income")
plt.title("predicted classes")

plt.tight_layout()
```

```python
from sklearn.metrics import adjusted_rand_score,normalized_mutual_info_score

ari=adjusted_rand_score(y,labels)
nmi=normalized_mutual_info_score(y,labels)

print(ari)
print(nmi)
```
