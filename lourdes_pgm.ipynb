```python
#Python program to print Even Numbers in a given range : 
```

```python
start=int(input("enter starting no:"))
end=int(input("enter ending no:"))
for i in range(start,end):
    if(i%2==0):
        print(i,",",end="")
```

```python
#Write a Python program to create a bar graph of the popularity of . Use multiple X values on the same
# chart for men and women. Sample Data: Means (men) = (22, 30, 35, 35, 26), Means (women) = (25, 32, 30,
# 35, 29)
```

```python
import numpy as np
import matplotlib.pyplot as plt
y1=[22,30,35,35,26]
y2=[25,32,30,35,29]
x_labels=['g1','g2','g3','g4','g5']
x1=np.arange(5)
width=0.40
plt.bar(x1-0.2,y1,color='green',width=width,label='men')
plt.bar(x1+0.2,y2,color='red',width=width,label='women')
plt.xticks(x1,x_labels)
plt.xlabel('groups')
plt.ylabel('scores')
plt.legend()
plt.show()
```
