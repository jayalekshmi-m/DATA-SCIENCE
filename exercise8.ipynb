```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#load iris dataset
iris=datasets.load_iris()
x=iris.data[:,:2]
y=iris.target

#create and train k means model
kmeans=KMeans(n_clusters=3,random_state=21)
kmeans.fit(x)
labels=kmeans.labels_

#plotting
plt.figure(figsize=(10,4))

#first subplot: actual species
plt.subplot(1,2,1)
plt.scatter(x[:,0],x[:,1],c=y,cmap="prism",edgecolor="k")
plt.title("actual")
plt.xlabel("sepal length")
plt.ylabel("sepal width")

#second subplot: predicted clusters
plt.subplot(1,2,2)
plt.scatter(x[:,0],x[:,1],c=labels,cmap="jet",edgecolor="k")
plt.title("predicted")
plt.xlabel("sepal length")
plt.ylabel("sepal width")

#layout ans show
plt.tight_layout()
```

```python
#explanation for first subplot:

# plt.subplot(1,2,1)  create the first subplot in a 1 row  2 column layout
# x[:,0], x[:,1]  sepal length vs sepal width
# c=y  color points based on true species labels
# cmap='prism'  color map for distinct colors
# edgecolor='k'  black border for each point
#  This shows the actual species grouping.

#explantion for second subplot:

# c=labels  color points according to cluster assigned by K-Means
# cmap='jet'  color map for clusters
#  This shows the unsupervised clustering results.
```

```python
print("Cluster Centers:\n", kmeans.cluster_centers_)
```

```python
# Second subplot - K-Means Predicted Clusters
plt.subplot(1,2,2)
plt.scatter(x[:,0], x[:,1], c=labels, cmap="jet", edgecolor="k")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, marker='*', color="black")  #  plotting centers
plt.title("K-Means Predicted Clusters")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```

```python
from sklearn.metrics import normalized_mutual_info_score,adjusted_rand_score

ari=adjusted_rand_score(y,labels)
nmi=normalized_mutual_info_score(y,labels)

print(ari)
print(nmi)
```
