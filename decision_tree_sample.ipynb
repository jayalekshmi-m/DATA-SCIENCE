```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
#1 create dataset
# Step 1: Create dataset
data = {
    'sore_throat': [1,0,1,1,0,0,1,0,1],
    'fever':       [1,0,0,0,1,0,0,1,1],
    'swollen_glands': [1,0,0,0,0,0,0,0,0],
    'congestion':  [1,1,1,0,1,1,0,1,1],
    'headache':    [1,1,0,0,0,0,0,1,1],
    'diagnosis':   ['Strep_Throat','Allergy','Cold','Strep_Throat','Cold','Allergy','Strep_Throat','Allergy','Cold']
}

df = pd.DataFrame(data)
print(df)
#This creates a small table (DataFrame) in memory that matches exactly what you see in the image.
```

```python
#2 Split into Features (X) and Target (y)
X = df[['sore_throat','fever','swollen_glands','congestion','headache']]
y = df['diagnosis']
#X  the symptom columns (input features)
#y  the output class (diagnosis)


#3 train a decision tree
clf = DecisionTreeClassifier(criterion="entropy")
clf.fit(X, y)
#criterion="entropy"  uses information gain to split data (like you did in the iris example).
#.fit()  trains the model.


#4 visualize the tree
plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, rounded=True, feature_names=X.columns, class_names=clf.classes_)
plt.title("Decision Tree for Diagnosis")
plt.show()


#5 predict new case
sample = [[0, 0, 1, 0, 0]]
prediction = clf.predict(sample)
print("Predicted diagnosis:", prediction[0])
```

```python
#easy way of above ques

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sample3.csv")

# Encode all columns at once
le = LabelEncoder()
df = df.apply(le.fit_transform)

# Split into X (features) and y (target)
X = df.iloc[:, :-1]   # all columns except last
y = df.iloc[:, -1]    # last column only


x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

model = DecisionTreeClassifier(criterion='entropy')
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

plt.figure(figsize=(12,8))
plot_tree(model, filled=True, rounded=True, feature_names=X.columns)
plt.show()

#  Predict a new case (much easier now)
new = [[0, 1, 1]]   # For example after encoding: Young=0, Medium=1, Yes=1
print("Prediction:", model.predict(new))
final=model.predict(new)
print("prediction:",le.inverse_transform(final)[0])
```
