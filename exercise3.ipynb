```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,6,18])
y=np.array([3,10,12,20])
plt.plot(x,y,color='r',marker=".",mfc="g",mec="g",linestyle=":")
```

```python
temp=np.array([12,14,16,18,20,22,24])
sales=np.array([100,200,250,400,300,450,500])
plt.plot(temp,sales,marker=".")
plt.xlabel("temperature")
plt.ylabel("sales")
```

```python
x_vals=[]
y_vals=[]
with open ("data.txt","r") as file:
    for line in file:
        x,y=map(float,line.strip().split())
        x_vals.append(x)
        y_vals.append(y)
plt.plot(x_vals,y_vals,marker="o",linestyle="-",color="blue")
```

```python
x=[1,2,3,4,5]
y1=[2,3,5,7,11]
y2=[1,4,6,8,10]
plt.plot(x,y1,label="prime growth",color="blue",marker="o")
plt.plot(x,y2,label="linear growth",color="green",linestyle="--")
plt.xlabel("time")
plt.ylabel("values")
plt.legend()
plt.show()
```

```python
a=np.array([10,20,15])
b=np.array([22,33,44])
plt.subplot(1,2,1)
plt.plot(a,b,marker=".")
plt.subplot(1,2,2)
plt.plot(b,a,marker="+")
```

```python
languages=np.array(["java","python","php","javascipt","c","c++"])
popularity=np.array([222,17.6,20,77,6.7,50])
plt.bar(languages,popularity)
plt.title("programming languages popularity")
```

```python
plt.barh(languages,popularity,color="red")
```

```python
colors=np.array(["blue","green","orange","purple","red","magenta"])
plt.bar(languages,popularity,color=colors)
```

```python
men=np.array([22,30,35,35,26])
women=np.array([25,32,30,35,29])
labels=np.array(["a","b","c","d","e"])
x=np.arange(len(labels))
width=0.25
fig,ax=plt.subplots()
rects1=ax.bar(x-width/2,men,width,label="men",color="powderblue")
rects2=ax.bar(x+width/2,women,width,label="women",color="teal")
ax.set_xlabel("category")
ax.set_ylabel("scores")
ax.set_title("scores by category and gender")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```

```python
ar1=np.array(["java","pyhton","php","javascript","c#","c++"])
ar2=np.array([22.2,17.6,8.8,8,77,6.7])
plt.pie(ar2, labels=ar1)
plt.axis("equal")
```

```python
import pandas as pd
df=pd.read_csv("medals.csv")
lables=df["country"]
sizes=df['gold_medals']
plt.pie(sizes,labels=lables)
plt.axis("equal")
plt.title("pie chart from csv")
```
